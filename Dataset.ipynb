```python
import pandas as pd

class_df = pd.read_csv(class_labels, header=None)
class_df = class_df[class_df[1].isin(classes)]
```

```python
train_df = pd.read_csv(train_annotations)
test_df = pd.read_csv(test_annotations)
validation_df = pd.read_csv(validation_annotations)
```

```python
train_df = train_df[train_df['LabelName'].isin(class_df[0])]
test_df = test_df[test_df['LabelName'].isin(class_df[0])]
validation_df = validation_df[validation_df['LabelName'].isin(class_df[0])]
```

```python
import os
from concurrent.futures import ThreadPoolExecutor

def process_file(class_label_dir, class_index, row):
    image_id = row['ImageID']
    file_name = f"{image_id}.txt"
    file_path = os.path.join(class_label_dir, file_name)

    if os.path.exists(file_path):
        x1, y1, x2, y2 = row[['XMin', 'YMin', 'XMax', 'YMax']].values
        x_center = (x1 + x2) / 2
        y_center = (y1 + y2) / 2
        width = x2 - x1
        height = y2 - y1

        with open(file_path, 'w') as f:
            f.write(f'{class_index} {x_center} {y_center} {width} {height}')


executor = ThreadPoolExecutor(max_workers=10)

for class_name in classes:
    class_index = classes_index[class_name]

    for dir, df in zip([train_dir, test_dir, validation_dir], [train_df, test_df, validation_df]):
        print(f'Processing {class_name} in {dir}')
        class_label_dir = os.path.join(dir, class_name, 'Label')
        for _, row in df.iterrows():
            executor.submit(process_file, class_label_dir, class_index, row)

executor.shutdown()
```

```python
import os
import xml.etree.ElementTree as ET

def convert_annotation(xml_file_path, txt_file_path):
    tree = ET.parse(xml_file_path)
    root = tree.getroot()

    with open(txt_file_path, 'w') as f:
        for obj in root.iter('object'):
            # difficult = obj.find('difficult').text
            cls = obj.find('name').text

            if cls not in classes_index:
                # Add new classes to classes_index dictionary
                if cls not in classes_index:
                    classes_index[cls] = len(classes_index)


            cls_id = classes_index[cls]
            xmlbox = obj.find('bndbox')
            x1, x2, y1, y2 = (int(xmlbox.find('xmin').text), int(xmlbox.find('xmax').text), int(xmlbox.find('ymin').text), int(xmlbox.find('ymax').text))
            width = int(root.find('size').find('width').text)
            height = int(root.find('size').find('height').text)

            x_center = (x1 + x2) / 2 / width
            y_center = (y1 + y2) / 2 / height
            w = (x2 - x1) / width
            h = (y2 - y1) / height

            f.write(f'{cls_id} {x_center} {y_center} {w} {h}\n')
```

```python
import os
import random
import shutil

def split_dataset(dir, images_dir, labels_dir, train_ratio=0.8, test_ratio=0.1, validation_ratio=0.1):
    images_dir = os.path.join(dir, images_dir)
    labels_dir = os.path.join(dir, labels_dir)

    # Create train, test and validation directories
    train_dir = os.path.join(dir, 'train')
    test_dir = os.path.join(dir, 'test')
    validation_dir = os.path.join(dir, 'validation')

    if not os.path.exists(train_dir):
        os.makedirs(train_dir)

    if not os.path.exists(test_dir):
        os.makedirs(test_dir)

    if not os.path.exists(validation_dir):
        os.makedirs(validation_dir)

    # Make list of names of all images without extension
    images_names = [file_name.split('.')[0] for file_name in os.listdir(images_dir)]

    # Shuffle the images names
    random.shuffle(images_names)

    # Split the images names into train, test and validation
    train_images_names = images_names[:int(train_ratio * len(images_names))]
    test_images_names = images_names[int(train_ratio * len(images_names)):int((train_ratio + test_ratio) * len(images_names))]
    validation_images_names = images_names[int((train_ratio + test_ratio) * len(images_names)):]

    # Copy images and labels to train, test and validation directories
    for image_name in train_images_names:
        shutil.copy(os.path.join(images_dir, image_name + '.jpg'), os.path.join(train_dir, image_name + '.jpg'))
        shutil.copy(os.path.join(labels_dir, image_name + '.txt'), os.path.join(train_dir, image_name + '.txt'))

    for image_name in test_images_names:
        shutil.copy(os.path.join(images_dir, image_name + '.jpg'), os.path.join(test_dir, image_name + '.jpg'))
        shutil.copy(os.path.join(labels_dir, image_name + '.txt'), os.path.join(test_dir, image_name + '.txt'))

    for image_name in validation_images_names:
        shutil.copy(os.path.join(images_dir, image_name + '.jpg'), os.path.join(validation_dir, image_name + '.jpg'))
        shutil.copy(os.path.join(labels_dir, image_name + '.txt'), os.path.join(validation_dir, image_name + '.txt'))
```

```python
import cv2
import matplotlib.pyplot as plt


def show_image_with_boxes(image_path, annotations_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (512, 512))

    with open(annotations_path, 'r') as f:
        for line in f.readlines():
            cls_id, x_center, y_center, w, h = line.strip().split(' ')
            cls_id = int(cls_id)
            x_center = float(x_center)
            y_center = float(y_center)
            w = float(w)
            h = float(h)

            x1 = int((x_center - w / 2) * 512)
            y1 = int((y_center - h / 2) * 512)
            x2 = int((x_center + w / 2) * 512)
            y2 = int((y_center + h / 2) * 512)

            image = cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```
